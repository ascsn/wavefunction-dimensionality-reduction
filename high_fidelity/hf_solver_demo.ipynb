```python
import numpy as np
from hf_solver import hf_solver
```

```python
# Initalizes the object
#  (h_bar, mass, ang_freq, q, dx, x_limit, dt, t_final)
sim = hf_solver(1, 1, 1, -5, 0.05, 5, 0.01, 5)

domain, timeframe = sim.get_ranges()

# Sets the initial state
sim.set_initial_state(x0 = 0.5)

# Sets the hamiltonian matrix and choose wavefunction scheme
# QHO: Quantum Harmonic Oscillator; GP: Gross-Pitaevskii
sim.construct_hamiltonian("GP")

# Runs the animation
sim.run_solver(gif = False, animation = True, collect_data = True)
```

```python
# Basic example of the run_simulation method, this particular instance will record data for the first 100 timesteps. Outputs are the final state, all saved states, all saved energies, and the total time taken.
# Initalizes the object
#  (h_bar, mass, ang_freq, dx, x_limit, dt, t_final)
sim = hf_solver(1, 1, 1, 0.05, 5, 0.01, 5)

# Sets the potential (currently hardcoded to quantum harmonic oscillator)
domain, timeframe = sim.get_ranges()

# Sets the initial state, in this case a guassian wave packet centered at zero. This will eventually all be done strictly through methods.
#initial_state_ = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * domain**2)
initial_state = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * domain**2)
sim.set_initial_state(initial_state)

# Sets the hamiltonian matrix
sim.construct_hamiltonian("GP")

# Runs the animation

f_state, a_states, energies, total_time = sim.run_simulation(100)
print(total_time, 'sec',
      a_states)
```
