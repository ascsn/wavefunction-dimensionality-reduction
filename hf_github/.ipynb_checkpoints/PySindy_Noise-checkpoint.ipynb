```python
import pysindy as ps
import numpy as np
from CAE_NIF import CAE
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
```

## Creating Equation and Explicit Derivative

```python
time_arr = np.arange(0,40,0.01)
X = np.column_stack([-4.35*(np.cos(2.0025*time_arr-3.6) - 1.6*np.cos(1.00125*time_arr-1.8))+6.7,-6*(np.cos(2*time_arr+3) - 0.6*np.cos(1*time_arr+1.5))+5 ])
Xderiv = np.concatenate(([np.zeros(2)], np.diff(X, axis=0) / (time_arr[1]-time_arr[0])))
Xderivexact = np.column_stack([4.35*(2.0025*np.sin(2.0025*time_arr-3.6) - 1.60125*np.sin(1.00125*time_arr-1.8)),12*np.sin(2*time_arr+3)-3.6*np.sin(time_arr+1.5)])


X_orig =  np.column_stack([X,Xderiv])
X_orig_exact =  np.column_stack([X,Xderivexact])
X_orig = X_orig_exact
```

## PySINDy

```python
library_functions = [lambda x: x, lambda x: x * x, lambda x, y: x * y]
library_function_names = [lambda x: x, lambda x:x+'^2', lambda x, y: x + y]
x_dot_library_functions = [lambda x: x]


deriv = np.concatenate(([np.zeros(2)], np.diff(X, axis=0) / (time_arr[1]-time_arr[0])))
plt.plot(X[1:,0], deriv[1:,0])
deriv2 = np.concatenate(([np.zeros(2)], np.diff(deriv, axis=0) / (time_arr[1]-time_arr[0])))
deriv3 = np.concatenate(([np.zeros(2)], np.diff(deriv2, axis=0) / (time_arr[1]-time_arr[0])))

X_ = np.column_stack([X_orig_exact])
ode_lib = ps.WeakPDELibrary(
    library_functions=library_functions,
    function_names=library_function_names,
    derivative_order=2,
    #feature_library=ps.PolynomialLibrary(degree=2,include_bias=False),
    spatiotemporal_grid=time_arr[1:-1],
    #spatial_grid=X,
    is_uniform=True,
    K=50,
)


# Instantiate and fit the SINDy model with the integral of u_dot
optimizer = ps.SR3(
    threshold=5, thresholder="l1", max_iter=500, normalize_columns=True, tol=1e-12
)
model = ps.SINDy(differentiation_method=ps.FiniteDifference(order=4), feature_library=ps.PolynomialLibrary(degree=1), 
        optimizer=ps.STLSQ(threshold=0.1,normalize_columns=False))

#model = ps.SINDy(feature_library=ode_lib, optimizer=ps.STLSQ(threshold=0.01,normalize_columns=False))
model.fit(X_[0:], t=time_arr[1]-time_arr[0])
model.print()
```

```python
x_simulated = model.simulate(X_[0], time_arr[0:], integrator_kws={'atol': 1e-12, 'method': 'RK45', 'rtol': 1e-12}, interpolator_kws={})

print(len(x_simulated))
plt.plot(time_arr[0:],x_simulated[:,0], label='Simulated')
plt.plot(time_arr[0:],X_[:,0], alpha=0.5, label='Actual')
plt.legend()
plt.show()
```

As it can be seen, Sindy is unable to get the correct dynamics when noise is at 0.01. We will now see if it can by using fourier to get the general dynamics and see how far that can go. When the time step was 0.001 Sindy was unable to get an dynamics with a noise of 0.01. However, after changing the time step to 0.01 it could.

```python
import numpy as np
import matplotlib.pyplot as plt

eq = []
for i in range(X.shape[1]):
    print(i)
    # Generate a sample signal
    #t = np.linspace(0, 40, 40000, endpoint=False)  # time points from 0 to 40 (exclusive) with 4000 samples
    signal = noisy_array[:,i]  # Example signal (replace with your data)
    
    # Compute Fourier transform
    fourier_transform = np.fft.fft(signal)
    freqs = np.fft.fftfreq(len(signal), time_arr[1] - time_arr[0])  # compute frequency axis
    
    # Example: Zeroing out frequencies above 'num' Hz
    num = 0.7
    fourier_transform[np.abs(freqs) > num] = 0
    
    # Extract amplitudes and phases
    amplitudes = np.abs(fourier_transform)
    phases = np.angle(fourier_transform)
    
    # Construct reconstructed signal
    #t_large = np.linspace(0,40,40000)
    waveform = np.zeros_like(time_arr)
    for i in range(len(freqs)):
        amplitude = amplitudes[i]
        phase = phases[i]
        freq = freqs[i]
        waveform += amplitude * np.cos(2 * np.pi * freq * time_arr + phase)/time_arr.shape[0]
    eq.append(waveform)

equation = np.array(eq).T
# Plotting
plt.figure(figsize=(12, 10))

# Original and Reconstructed signal plot
plt.subplot(3, 1, 1)
plt.plot(time_arr[100:], signal[100:], label='Original Signal')
plt.plot(time_arr[100:], waveform[100:], label='Reconstructed Signal', linestyle='--')
plt.title('Original and Reconstructed Signals')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

# Fourier transform plot
plt.subplot(3, 1, 2)
plt.scatter(freqs, np.abs(fourier_transform))
plt.title('Modified Fourier Transform')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.xlim(-num-1, num+1)

# Phase plot
plt.subplot(3, 1, 3)
plt.plot(freqs, phases)
plt.title('Phases of Fourier Coefficients')
plt.xlabel('Frequency')
plt.ylabel('Phase (radians)')
plt.tight_layout()
plt.show()
```

#### PySINDy with more advance finite difference method

```python
def finite_difff(points, time):
    t = time
    X=points
    finite_diff = np.zeros_like(X)
    for i in range(finite_diff.shape[1]):
        for j in range(finite_diff.shape[0]):
            if j < 4:
                finite_diff[j,i] = (-25/12*X[j,i] + 4*X[j+1,i] -3*X[j+2,i] + 4/3*X[j+3,i] - 1/4*X[j+4,i])/(t[1]-t[0])
            elif j > finite_diff.shape[0]-4:
                finite_diff[j,i] = (25/12*X[j,i] - 4*X[j-1,i] +3*X[j-2,i] - 4/3*X[j-3,i] + 1/4*X[j-4,i])/(t[1]-t[0])
            else:
                finite_diff[j,i] = (-1/60*X[j-3,i] + 3/20*X[j-2,i] -3/4*X[j-1,i] + 3/4*X[j+1,i] - 3/20*X[j+2,i] + 1/60*X[j+3,i])/(t[1]-t[0])
    return finite_diff
```

```python
first_deriv = finite_difff(equation,time_arr)
second_deriv = finite_difff(first_deriv,time_arr)
X_ = np.column_stack([equation,first_deriv])
print(X_)
plt.plot(X[:,0],first_deriv[:,0])
ode_lib = ps.WeakPDELibrary(
    library_functions=library_functions,
    function_names=library_function_names,
    derivative_order=2,
    #feature_library=ps.PolynomialLibrary(degree=2,include_bias=False),
    spatiotemporal_grid=time_arr,
    #spatial_grid=X,
    is_uniform=True,
    K=50,
)


# Instantiate and fit the SINDy model with the integral of u_dot
optimizer = ps.SR3(
    threshold=5, thresholder="l1", max_iter=500, normalize_columns=True, tol=1e-12
)
model = ps.SINDy(differentiation_method=ps.FiniteDifference(order=4), feature_library=ps.PolynomialLibrary(degree=1), 
       optimizer=ps.STLSQ(threshold=0,normalize_columns=False))

#model = ps.SINDy(feature_library=ode_lib, optimizer=ps.STLSQ(threshold=0.01,normalize_columns=False))
model.fit(X_[100:-100], t=time_arr[1]-time_arr[0], ensemble=True, n_models=100, quiet=True)
model.print()
```

Above, it can be seen that this time by using the more accurate finite difference method we can predict starting at 0. Although this works, there could be some work arounds to be able to accurately predict starting at 0. The code below tries to take the middle section of the Fourier transform and pair it with 1. the normal noisy data and 2. the same data that has been smoothed.

```python
end=100
smooth = 0.001
t_new = time_arr[0:end]
spline1 = UnivariateSpline(t_new, noisy_array[:end,0],s=smooth)
spline2 = UnivariateSpline(t_new, noisy_array[:end,1],s=smooth)
l1 = spline1(t_new)
l2 = spline2(t_new)
lat_beg = np.column_stack([l1,l2])

t_end = time_arr[-end:]
spline1_end = UnivariateSpline(t_end, noisy_array[-end:,0],s=smooth)
spline2_end = UnivariateSpline(t_end, noisy_array[-end:,1],s=smooth)
l_end1 = spline1_end(t_end)
l_end2 = spline2_end(t_end)
lat2 = np.column_stack([l_end1,l_end2])
print(lat2.shape)
plt.plot(lat2[:,0], label='Smoothed Curve', color='r')
plt.plot(noisy_array[-end:,0])
#plt.xlim(0,60)
#plt.ylim(12,13)
print(lat2[0,1])
print(noisy_array[-end,1])
```

Keeping track of the differe between the enpoints are important as it can mess with derivatives if they are too far off.

```python
X_nfn =np.concatenate((noisy_array[:end], equation[end:-end],noisy_array[-end:] ))#nfn for noisy beginning, fourier midesection and noisy end
X_sfs = np.concatenate((lat_beg, equation[end:-end],lat2))  #sfs for the smooth, fourier and smooth
plt.plot(X_nfn[:,0])
print(X_nfn.shape)
plt.xlim(90,110)
```

```python
deriv_nfn = finite_difff(X_nfn,time_arr)
deriv_sfs = finite_difff(X_sfs,time_arr)
plt.subplot(2,1,1)
plt.plot(deriv_nfn[0:,0])

plt.subplot(2,1,2)
plt.plot(deriv_sfs[0:,0])
```

As it can be seen above there is a break in the graph in where we connected the arrays together. However, we can smooth the latent space once again to get rid of that.

```python
smooth = 0.1
spline1 = UnivariateSpline(time_arr, X_sfs[:,0],s=smooth)
spline2 = UnivariateSpline(time_arr, X_sfs[:,1],s=smooth)
l1 = spline1(time_arr)
l2 = spline2(time_arr)
smoothed_sfslatent = np.column_stack([l1,l2])

spline1 = UnivariateSpline(time_arr, X_nfn[:,0],s=smooth)
spline2 = UnivariateSpline(time_arr, X_nfn[:,1],s=smooth)
l1 = spline1(time_arr)
l2 = spline2(time_arr)
smoothed_nfnlatent = np.column_stack([l1,l2])

plt.plot(time_arr,smoothed_sfslatent[:,0], label='Smoothed edges')
plt.plot(time_arr,smoothed_nfnlatent[:,0], label='Noisy edges')
plt.plot(time_arr,noisy_array[:,0], label='Noisy')
plt.xlim(0,1)
plt.legend()
rmse = np.sqrt(np.mean((smoothed_sfslatent[:,0] - noisy_array[:,0])**2))
rmse2 = np.sqrt(np.mean((smoothed_nfnlatent[:,0] - noisy_array[:,0])**2))
print("RMSE of smooth ends:", rmse)
print("RMSE of noisy ends:", rmse2)
```
