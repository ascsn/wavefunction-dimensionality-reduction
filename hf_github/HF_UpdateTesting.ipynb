```python
import numpy as np
import matplotlib.pyplot as plt
from hf_latest import hf_solver
```

```python
''' 
Initalizes the object
Inputs: (h_bar, mass, ang_freq, q, dx, x_limit, dt, t_final)
'''
sim = hf_solver(1, 1, 1, -5, 0.05, 5, 0.05, 5)
''' 
Sets the initial state in the form of a guassian wavepacket
Inputs: (x0, sigma, p)
        x0: initial position 
        sigma: width of wave 
        p: momentum
'''
sim.set_initial_state(x0 = 0, p=1)
```

```python
''' 
Sets the hamiltonian matrix and choose wavefunction scheme
Inputs: (type)
        options:
        QHO: Quantum Harmonic Oscillator 
        GP: Gross-Pitaevskii
'''
sim.construct_hamiltonian("QHO")
```

```python
'''
Runs the solver
Inputs (snapshot, snapshot_freq)
'''
sim.run_solver() 
```

```python
'''
Visualize the wavefunction
Inputs: show: shows the movie in the iPython display 
        save: saves the movie locally as a MP4 
        fps: adjusts the frames per second on the MP4, default is 30)
'''
sim.animation(show = True, save = False)
```
